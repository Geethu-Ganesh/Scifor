```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()

x=df1[['Weight','Volume']]

scaledx=scale.fit_transform(x)
print(scaledx)
```

```python
scaled_column=pd.DataFrame(scaledx)
scaled_column
```

```python
df1=pd.concat([df1,scaled_column],axis=1)
df1.rename(columns={0:"weight_scaled",1:"volume_scaled"},inplace=True)
df1
```
